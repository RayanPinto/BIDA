```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("Mall_Customers - Mall_Customers.csv")
a=df.head()
print(a)
```

```python
f=df[['Age','Annual Income (k$)','Spending Score (1-100)']]
s=StandardScaler()
ss=s.fit_transform(f)
wcss=[]
for k in range(2,11):
    km=KMeans(n_clusters=k,n_init=10,random_state=42)
    km.fit(ss)
    wcss.append(km.inertia_)
plt.plot(range(2,11),wcss,marker='o')
plt.xlabel("number of clusters")
plt.ylabel("wcss")
plt.show()
```

```python
for k in range(2,11):
    km=KMeans(n_clusters=k,random_state=42)
    lb=km.fit_predict(ss)
    score=silhouette_score(ss,lb)
    print(f" K={k} and silhouette score is {score:.3f}")
opk=6
km=KMeans(n_clusters=opk,random_state=42)
df['cluster']=km.fit_predict(ss)
print(df.groupby('cluster')[['Age','Annual Income (k$)','Spending Score (1-100)']].mean())
plt.scatter(ss[:,0],ss[:,1],c=df['cluster'],cmap='viridis')
plt.xlabel("number of clusters")
plt.ylabel("wcss")
plt.show()
```

```python
fig =plt.figure(figsize=(8,6))
ax=fig.add_subplot(111,projection='3d')
ax.scatter(ss[:,0],ss[:,1],ss[:,2],c=df['cluster'],cmap='viridis')
plt.show()
```

```python
from mlxtend.frequent_patterns import apriori ,association_rules
from mlxtend.preprocessing import TransactionEncoder
df=pd.read_csv("Market_Basket_Optimisation - Market_Basket_Optimisation.csv")
tr=[]
for i in range(0,len(df)):
    tr.append([str(df.values[i,j]) for j in range(df.shape[1]) if str(df.values[i,j])!='nan'])
te=TransactionEncoder()
te_array=te.fit(tr).transform(tr)
encoded=pd.DataFrame(te_array,columns=te.columns_)
```

```python
freq_item=apriori(encoded,min_support=0.01,use_colnames=True)
print(freq_item.shape[0])
```

```python
rules=association_rules(freq_item,min_threshold=0.01,metric='confidence')
rules=rules[rules['antecedents'].apply(lambda x:  len(x)>=1) & rules['consequents'].apply(lambda x:  len(x)>=1)]
```
