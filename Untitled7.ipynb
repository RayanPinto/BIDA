```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv("Mall_Customers - Mall_Customers.csv")
df.head(5)
```

```python
feature=df[['Age','Annual Income (k$)','Spending Score (1-100)']]
```

```python
scaler=StandardScaler()
ss=scaler.fit_transform(feature)
wcss=[]
```

```python
for k in range(2,11):
    km=KMeans(n_clusters=k,random_state=42,n_init=10)
    
    km.fit(ss)
    wcss.append(km.inertia_)
    
```

```python
import matplotlib.pyplot as plt
plt.plot(range(2,11),wcss,marker='o')
plt.xlabel("number of clsuters")
plt.ylabel("wcss")
plt.show()
```

```python
for k in range(2,11):
    km=KMeans(n_clusters=k,random_state=42,n_init=10)
    
    lb=km.fit_predict(ss)
    sss=silhouette_score(ss,lb)
    print(sss)
```

```python
km=KMeans(n_clusters=6,random_state=42)
df['cluster']=km.fit_predict(ss)
print(df.groupby('cluster')[['Age','Annual Income (k$)','Spending Score (1-100)']].mean())
```

```python
plt.scatter(ss[:,0],ss[:,1],c=df['cluster'],cmap='viridis')
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("customer segements")
plt.show()
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori ,association_rules
```

```python
tr=[]
df=pd.read_csv("Market_Basket_Optimisation - Market_Basket_Optimisation.csv")
for i in range(len(df)):
    tr.append([str(df.values[i,j]) for j in range(df.shape[1]) if str(df.values[i,j])!='nan'])
    

```

```python
te=TransactionEncoder()
te_array=te.fit(tr).transform(tr)
encoded=pd.DataFrame(te_array,columns=te.columns_)
```

```python
freq_item=apriori(encoded,min_support=0.01,use_colnames=True)
```

```python
print(freq_item.shape[0])
```

```python
rules=association_rules(freq_item,metric='confidence',min_threshold=0.1)
```

```python
rules=rules[rules['antecedents'].apply(lambda x:  len(x)>=1) & rules['consequents'].apply(lambda x: len(x)>=1)]
```

```python
rules[['antecedents','consequents','support','confidence','lift']]
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

df=pd.read_csv("customer_review - customer_review.csv")
df=df.dropna(subset=["Review"])

df=df[df["Review"].str.strip() !='']
df['Review'] =df['Review'].astype(str)
x=df['Review']
y=df['Sentiment']

x_tr,x_te,y_tr,y_te=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)

tfid=TfidfVectorizer(stop_words='english',max_features=5000,lowercase=True)
xtrtf=tfid.fit_transform(x_tr)
xtetf=tfid.transform(x_te)

nb=MultinomialNB()

nb.fit(xtrtf,y_tr)
y_pred=nb.predict(xtetf)
```

```python
print(classification_report(y_te,y_pred))
```

```python
from bs4 import BeautifulSoup
from sklearn.decomposition import NMF
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
files=["Artificial intelligence/Artificialintelligence.html","Startups_TechCrunch/Startups_TechCrunch.html"]
```

```python
def extract(fp):
    with open(fp,'r',encoding="utf-8") as f:
        soup=BeautifulSoup(f.read(),"html.parser")
    for t in soup(["noscript","script","style"]):
        t.decompose
    return soup.get_text()
```

```python
docs=[extract(f) for f in files if len(extract(f))>200]
```

```python
vectorizer=TfidfVectorizer(stop_words='english',max_df=0.9,min_df=0.1,ngram_range=(1,2))
```

```python
x=vectorizer.fit_transform(docs)
terms=vectorizer.get_feature_names_out()
score=x.mean(axis=0).A1
```

```python
top_idx=score.argsort()[::-1][:25]
for i,idx in enumerate(top_idx):
    print(f" {terms[idx]} and {score[idx]}")
```

```python
n=min(6,len(docs))
nmf=NMF(n_components=n,random_state=42)
h=nmf.fit(x)
for i,topic in enumerate(nmf.components_):
    top_topic=topic.argsort()[::-1][:10]
    print(f" topic {i+1} : {terms[top_topic]}")
```

```python
import re,nltk,pandas 
from nltk.sentiment import SentimentIntensityAnalyzer

from bs4 import BeautifulSoup

files=r"r_news.html"
com=[]
soup=BeautifulSoup(open(files,encoding="utf-8",errors="ignore").read(),"lxml")
selectors=[".comment",".comment-body",".comment-content","s-porse",".post-message"]
for sel in selectors:
    for n in soup.select(sel):
        t=re.sub(r'https?://\S+|www\.\s+'," ",n.get_text(" ",strip=True))
        if 8<en(t)<=1000:
            com.append(t)
if not com:
    com=[ p.get_text(" ",strip=True) for p in soup.find_all("p")]
    com=[re.sub(r'https?://\S+|www\.\s+'," ",t) for t in com if len(t)>20]

nltk.download("vader_lexicon",quiet=True)
sia=SentimentIntensityAnalyzer()

def get_sentiment(t):
    c=sia.polarity_scores(t)["compound"]
    return "postive" if c>0.05 else ("negative" if c<-0.05 else "neutral")
    

df=pd.DataFrame({"comments":com})
df['sentiment']=df['comments'].apply(get_sentiment)

dist=df['sentiment'].value_counts()

overal=dist.idxmax() if not dist.empty else "neutral"
print(overal.upper())

print(dist)

plt.bar(dist.index,dist.values,color=["green","red","grey"])
plt.show()

```

```python
from bs4 import BeautifulSoup
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import re
# Files to process
files = ["The Indian Express/TheIndianExpress.html", "The Economic Times/TheEconomicTimes.html", "TIE/TIE.html"]
def extract_text(filepath):
    with open(filepath, 'r', encoding='utf-8') as f:
        soup = BeautifulSoup(f.read(), 'html.parser')
    for tag in soup(["script", "style"]):
        tag.decompose()
    txt="".join(p.get_text() for p in soup.find_all("p"))
    txt=re.sub(r'https:?//\S+|www\.\s+',' ',txt)
    txt=re.sub(r'\s+','',txt)
    return txt

txts=[ extract_text(f) for f in files if len(extract(f)) >20]

vectorizer=CountVectorizer(stop_words='english',max_df=0.95 ,min_df=2)
x=vectorizer.fit_transform(txts)
terms=vectorizer.get_feature_names_out()
score=x.mean(axis=0).A1
top_idx=score.argsort()[::-1][:25]
for i,ix in enumerate(top_idx):
    print(f" {terms[ix]} \n{score[ix]}")

k=min(6,len(txts))

lda=LatentDirichletAllocation(n_components=k,random_state=42)
lda.fit(x)
for i,topic in enumerate(lda.components_):
    top_topic=topic.argsort()[::-1][:10]
    print(f" {" ||".join(terms[top_topic])}")

```
